```python

student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 65,
    "Eve": 95
}
search= input("Enter the student's name to find their marks: ")
if search in student_marks:
    marks = student_marks[search]
    print(f"{search}'s marks: {marks}")
else:
    print(f"Student '{search}' not found in the records.")
```

```python
user_credentials = {
    "admin": "1234",
    "student": "abcd"
}
entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

if entered_username in user_credentials:
    if user_credentials[entered_username] == entered_password:
        print(f"Login successful! Welcome, {entered_username}!")
    else:
        print("Incorrect password. Please try again.")
else:
    print("Username not found. Please check your username and try again.")
```
